Dot Product
===========

Given an input vector and a neuron (weight) vector, the dot product operation measures the similarity between the two vectors.

The output value range is (- Infinity, Infinity).

If our neuron is [ 1, 1 ] and our input vector is [ 2, 2 ], the output value is 4: the two vectors are pointed in the same direction.

If our neuron is [ 1, 1 ] and our input vector is [ -2, -2 ], the output value is -4: the two vectors are pointed in opposite directions.

If we normalized the output value by dividing by the lengths of the two vectors, we would be computing the cosine similarity between the two vectors.

```python
import numpy as np
np.dot([ 1, 1 ], [ 2, 2 ])
```

```python
np.dot([ 1, 1 ], [ -2, -2 ])
```

```python
np.linalg.norm([ 2, 2 ])
```

```python
np.sqrt(2**2 + 2**2)
```

```python
np.dot([ 1, 1 ], [ 2, 2 ]) / (np.linalg.norm([ 1, 1 ]) * np.linalg.norm([ 2, 2 ]))
```

```python
np.dot([ 1, 1 ], [ -2, -2 ]) / (np.linalg.norm([ 1, 1 ]) * np.linalg.norm([ -2, -2 ]))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use("seaborn")
%matplotlib inline

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = slope * x_vals + intercept
    plt.plot(x_vals, y_vals, '-')

plt.title('Contour Lines for the Neuron [ 1, 1 ]\n(dot product value is same for all points on a line)')
axes = plt.gca()
axes.set_xlabel('Input Feature 0')
axes.set_ylabel('Input Feature 1')
plt.axis([ -3, 3, -3, 3 ])
for intercept in range(-6, 6 + 1):
    abline(-1, intercept)

plt.show()
```

Questions
---------
1. Which line would form a decision boundary for this neuron?
2. What is the effect of adding a bias to the dot product?

Answers
-------
1. The line defined by the endpoints [-3, 3] and [3, -3]; because the value of the dot product is zero.
2. It shifts the decision boundary up or down; e.g. using a bias of +1 would shift the decision boundary down (because -1 + 1 is zero).
